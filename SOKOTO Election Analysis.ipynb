# Neighbour Identification
Calculating the Geodistic distance and Identifying Neighboring Polling Units within a 10km radius.

```python
# Installing the geopy library in my Jupyter Notebook

!pip install geopy 
```

```python
# Import the geodesic function from the 'geopy.distance' module. 
# The function is used to calculate the distance between two geographical points.

from geopy.distance import geodesic
```

```python
# Define the radius.
# Sets the radius within which to find neighbors to 10 kilometer.

radius_km = 10
```

```python
# Initializing an empty list to store neighbours for each polling unit

neighbours = []
```

```python
# The function to calculate neighbours within the defined radius

def find_neighbours(data, radius):
    for index, row in data.iterrows():
        pu_location = (row['Latitude'], row['Longitude'])
        pu_neighbours = []
        for idx, neighbour_row in data.iterrows():
            if index != idx:
                neighbour_location = (neighbour_row['Latitude'], neighbour_row['Longitude'])
                distance = geodesic(pu_location, neighbour_location).kilometers
                if distance <= radius:
                    pu_neighbours.append(idx)
        neighbours.append(pu_neighbours)
    return neighbours
```

```python
# Find neighbours for each polling unit

neighbours = find_neighbours(Sokoto_data, radius_km)
```

```python
# Adding neighbours information to the dataset

Sokoto_data['Neighbours'] = neighbours
```

```python
# Counting the number of neighbours for each polling unit

Sokoto_data['Neighbours_Count'] = Sokoto_data['Neighbours'].apply(len)
```

# Outlier Score Calculation
Comparing the votes each party received with those of its neighbouring units.

```python
# Function to calculate outlier scores

def calculate_outlier_scores(data):
    # Initialize outlier score columns
    data['APC_Outlier_Score'] = 0
    data['LP_Outlier_Score'] = 0
    data['PDP_Outlier_Score'] = 0
    data['NNPP_Outlier_Score'] = 0
    
    for index, row in data.iterrows():
        neighbours = row['Neighbours']
        if neighbours:  # Check if there are any neighbours
            # Compute mean votes for each party among neighbours
            avg_apc = data.loc[neighbours, 'APC'].mean()
            avg_lp = data.loc[neighbours, 'LP'].mean()
            avg_pdp = data.loc[neighbours, 'PDP'].mean()
            avg_nnpp = data.loc[neighbours, 'NNPP'].mean()
            
            # Calculate outlier scores
            data.at[index, 'APC_Outlier_Score'] = abs(row['APC'] - avg_apc)
            data.at[index, 'LP_Outlier_Score'] = abs(row['LP'] - avg_lp)
            data.at[index, 'PDP_Outlier_Score'] = abs(row['PDP'] - avg_pdp)
            data.at[index, 'NNPP_Outlier_Score'] = abs(row['NNPP'] - avg_nnpp)
        else:  # If no neighbours, outlier score is set to NaN to indicate no comparison
            data.at[index, 'APC_Outlier_Score'] = np.nan
            data.at[index, 'LP_Outlier_Score'] = np.nan
            data.at[index, 'PDP_Outlier_Score'] = np.nan
            data.at[index, 'NNPP_Outlier_Score'] = np.nan
            
    return data
```
